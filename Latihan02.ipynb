```python
#Spelling correction menggunakan python
!pip install pyspellchecker
from spellchecker import SpellChecker

spell = SpellChecker()

#input teks dari user
text = input("Masukkan Teks: ")

#cek spelling dari input teks
corrected_text = spell.correction(text)

# Cek apakah teks yang dikoreksi tidak None dan tidak sama dengan teks asli
if corrected_text and corrected_text != text:
    print("Teks yang dikoreksi:", corrected_text)
else:
    print("Teks sudah benar atau tidak ada saran koreksi yang sesuai.")
```
